```python
# using while loop in list and dictionaries

unconfirmed_user = ['alex','elice','david','lamar','travis']
confirmed_user = []

while unconfirmed_user:
    name = unconfirmed_user.pop()

    print(f'verifying user: {name} ......')
    confirmed_user.append(name.title())

print('\nthe following have been confirmed:\n')

for name in confirmed_user:    
    print(name)
```

```python
# removing all instance in a list in one go

pets = ['cat','cat','cat','dog','fish','cow','horse','parrot','pigeon']

while 'cat' in pets:
    pets.remove('cat')
print(pets)    
```

```python
# filling the dict with user response

response = {}
active = True

while active:
    name = input('whats your name')
    mountain = input('mountain you wanna climb')
    response[name] = mountain

    repeat = input('go further yes or no')
    if repeat == 'no':
        active = False

for i,j in response.items():
    print(f'{i} would like to climb {j}')        
```

```python
sandwiches = [
    "Club Sandwich",
    "Grilled Cheese",
    "Peanut Butter & Jelly",
    "Philly Cheese",
    "Reuben",
    "Sloppy Joe",
    "Tuna Melt",
    "Tuna Melt",
    "Tuna Melt",
]

finished_sandwich = []

print('Deli has run out of Tuna Melt.\n')

# Remove all "Tuna Melt" from the list
while "Tuna Melt" in sandwiches:
    sandwiches.remove("Tuna Melt")

# Process remaining orders
while sandwiches:
    order = sandwiches.pop()
    print(f"I made {order}.")
    finished_sandwich.append(order)


```

```python
print(finished_sandwich[::-1])
```

```python
# 7-10

poll = {}

while True:
    name = input('enter your name')
    place = input('enter your dream destination')
    poll[name] = place
    message = input('yes/no')
    
    if message == 'no':
        break

for name,place in poll.items():
    print(f'{name} want to go {place}')
```
