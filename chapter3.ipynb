```python
# modifying adding or removing 

motercycles = ['honda', 'kawasaki', 'yamaha', 'suzuki', 'harley', 'ktm']

motercycles[2] = 'ducati'

print(motercycles)
```

```python
# trying with empty list

cars = []
cars.append('porche')
cars.append('Audi')
cars.append('Bmw')
cars.append('Mercedes')
cars.append('Range Rover')
cars.append('Ferrari')
print(cars)
```

```python
# insert any element at any position 
# every time you run it add a value

motercycles.insert(3,'hasqverna')
print(motercycles)
```

```python
# removing the element
# every time you run it delete a value 
del motercycles[3]
print(motercycles)
```

```python
# removing it by pop method which works like removing top of the stack

pop_element = motercycles.pop()
print(motercycles)
print(pop_element)
```

```python
last_owned_motercycle = pop_element
print(f'the last motercycle i bought this year is : {last_owned_motercycle.title()}.')
```

```python
# poping item form any position 

first_motercycle = motercycles.pop(0)
print(f'my first motercycle i owned was {first_motercycle.title()}.')
```

```python
# removing an item by value

print(motercycles)

too_expensive = 'ducati'

motercycles.remove(too_expensive)
print(motercycles)

print(f'\n A {too_expensive.title()} is to expensive for me ')
```

```python
# organizing the list

#sorting list permanently with sort()

# sort() is permanent 

# sorted() is temperary
print('original  : ',cars)

cars.sort()
print(cars)

cars.sort(reverse=True)
print(cars)
```

```python
alphabets = ['d','e','r','y','j','w','a','s','o','g']

print(alphabets) # original

print(sorted(alphabets)) # it sort but not permanent change

print(alphabets) # original 
```

```python
countries = ['Italy','New Zealand','Greece','Argentina','Spain','Australia']

print('original one : ' ,countries)

print('sorted without real list changing change :', sorted(countries))

print('same as original one : ' ,countries)

print('sorted in reverse without real list changing :', sorted(countries,reverse=True))

print('same as original another time: ' ,countries)

```

```python
print('original : ',countries)

countries.reverse() # changing order not sorting

print('order changed : ',countries)

countries.reverse()

print('back in normal original state : ',countries)
```

```python
print(countries)

countries.sort()

print(countries)
```

```python
print(countries)

countries.sort(reverse=True)

print(countries)
```
