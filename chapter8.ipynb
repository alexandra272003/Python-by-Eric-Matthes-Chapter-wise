```python
# equivalent function calls positional,keyword,defaulf 

def pets(pet_name,animal_type='dog'): #default value dedi humne phele se
    '''displays animal type and animal name'''
    print(f'i have a {animal_type.title()}.')
    print(f'my {animal_type.title()}\'s name is {pet_name.title()}.\n')

pets('gabbar')

pets(pet_name='gabbar')

pets(pet_name='yakuza',animal_type='horse') # if i change the value which > than default value 

```

âœ… pets(pet_name='gabbar', animal_type='horse')
Now you overwrote the default by explicitly passing 'horse' as the animal_type.

What happens here:

The default 'dog' is ignored because you provided a new value.

This is exactly how default values work: they're only used if nothing is provided.

```python
# making an argument optional

def name_optional(first_name, last_name, middle_name=''):
    '''Return optional full name with or without middle name'''
    if middle_name:
        full_name = f'{first_name} {middle_name} {last_name}'
    else:
        full_name = f'{first_name} {last_name}'
    return full_name.title()

data_scientist = name_optional('alexandra', 'pratap', 'singh')
musician = name_optional('aditya', 'sisodia')

print(musician)         # Aditya Sisodia
print(data_scientist)   # Alexandra Singh Pratap

    
```

```python
def build_person(first_name,last_name,age=None):
    '''return info about person'''
    person = {'first': first_name , 'last': last_name}
    
    if age:
        person['age'] = age
    return person

musician = build_person('alexandra','singh',age=27)
print(musician)
musician = build_person('kendrik','leamar')
print(musician)
```

```python
# modifying a list of a function

def print_model(unprinted_design,completed_design):
    while unprinted_design:
        current_model = unprinted_design.pop()
        print(f'Printing model : {current_model}')
        completed_design.append(current_model)

def show_model(completed_design):
    print('\nThe following model have been printed:\n')
    for model in completed_design:
        print(model)

unprinted_design = ['phone case','transformers','ninja hattori','dragons']
completed_design = []

print_model(unprinted_design,completed_design)
show_model(completed_design)
```

```python
# 8-10
def show_message(texts):
    '''Function takes messages and prints them'''
    for text in texts:
        print(f'{text}')

def send_message(texts, sent_message):
    for i in range(len(texts)):
        message = texts.pop()
        sent_message.append(message)

texts = ['hello', 'hii', 'hola', 'namaste']
sent_message = []

# First show original messages
print("Original texts:")
show_message(texts)

# Send messages (from a copy)
send_message(texts[:], sent_message)

# Show messages after sending
print("\nAfter send_message() call:")
print("texts (original list):")
show_message(texts)  # should remain unchanged

print("sent_message (new list):")
show_message(sent_message)  # should contain all moved messages

```

```python
def make_pizza(*toppings):
    '''summarize the pizza we are about to make'''
    print('\nmaking a pizza with the following toppings:')
    for toppings in toppings:
        print('-',toppings)

make_pizza('cheese','pepporoni','green peppers')     
make_pizza('cheese','pepporoni') 
```

```python
# mixing positional and arbitrary arguments

def make_pizza(size,*toppings):
    '''summrize pizza that we are about to make'''
    print(f'\nMaking {size}-inch pizza with following toppings')
    for topping in toppings:
        print('-',topping)

make_pizza(12,'cheese','pepporoni','green peppers')     
make_pizza(16,'cheese','pepporoni','pineapple')         
```
