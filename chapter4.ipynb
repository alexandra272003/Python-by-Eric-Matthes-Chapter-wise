WORKING WITH PART OF LIST

```python
# SLICING A LIST

players = ["Alice", "Bob", "Charlie", "David", "Eve","Alia", "Boby", "Charl", "Dave", "Eveaa"]

print(players[0:3])

print(players[1:4])

print(players[:4])

print(players[2:])

print(players[-3:]) # -ve indexing works

print(players[::5]) # jumping by giving 3rd argument
```

```python
print("Here are the first three players team :")

for player in players[:3]:
    print(player.title ())
```

```python
# makeing copy of my list

my_foods = ["Pizza", "Burger", "Pasta", "Sushi", "Salad"]
friend_food = my_foods[:]

print("here's my favourite food : ",my_foods)
print("here's my friends favourite food : ",friend_food)
```

```python
# prove that given lists my food and friend food are two distinct list


my_foods.append('ice cream')
friend_food.append('falafal')

print(my_foods)
print(friend_food)
```

```python
my_foods = ["Pizza", "Burger", "Pasta", "Sushi", "Salad"]
friend_food = my_foods

print("here's my favourite food : ",my_foods)
print("here's my friends favourite food : ",friend_food)

'''tht is same list cuz we are assigning the variable friend food to my food we are not copying the list like above'''
```
