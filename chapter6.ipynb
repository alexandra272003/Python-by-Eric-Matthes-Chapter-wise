```python
#changing 1 to 5 alien characteristics

for alien in aliens[:5]:
    if alien['color'] == 'green':
        alien['color'] = 'yellow'
        alien['point'] = 10
        alien['speed'] = 'medium'

    elif alien['color'] == 'yellow':
        alien['color'] = 'red'
        alien['point'] = 15
        alien['speed'] = 'fast'

    
    elif alien['color'] == 'red':
        alien['color'] = 'green'
        alien['point'] = 20
        alien['speed'] = 'slow'    


for i in aliens:
    print(i)
```

```python
# a dict in a dict

users = {
    'sarah': {
        'first_name': 'Sarah',
        'last_name': 'Connor',
        'location': 'Los Angeles'
    },
    'alex': {
        'first_name': 'Alex',
        'last_name': 'Mercer',
        'location': 'New York'
    },
    'emma': {
        'first_name': 'Emma',
        'last_name': 'Watson',
        'location': 'London'
    }
}

for user,info in users.items():
    print(f'\nusername {user}:')
    full_name = f'{info['first_name']} {info['last_name']}'
    location = f'{info['location']}'

    print(f'Full Name: {full_name.title()}')
    print(f'Location: {location.title()}')
```

```python
# 6-11

cities = {
    'new_york': {
        'country': 'USA',
        'population': 8_500_000,
        'fac': 'Known as the city that never sleeps.'
    },
    'tokyo': {
        'country': 'Japan',
        'population': 13_960_000,
        'fac': 'Home to the busiest train station in the world (Shinjuku).'
    },
    'paris': {
        'country': 'France',
        'population': 2_160_000,
        'fac': 'The Eiffel Tower was originally intended to be dismantled after 20 years.'
    },
    'sydney': {
        'country': 'Australia',
        'population': 5_230_000,
        'fac': 'Famous for its Opera House and Harbour Bridge.'
    }
}


for city, city_info in cities.items():
    print(f"\n{city.title()} and its information is:")

    country = city_info['country']
    population = city_info['population']
    fact = city_info['fac']

    print(f'Country: {country}')
    print(f'Population: {population}')
    print(f'Fun Fact: {fact}')

```
